```python
#importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
#Cargar archivo csv desde seaborn
df_inicial=pd.read_csv('valencia_trabajo.csv')
display(df_inicial.head(5))
```

```python
#Eliminar Unnamed: 0
Valencia = df_inicial
Valencia
```

```python
#Formato de host_id, latitude, longitude y host_is_superhost
Valencia['host_id'] = Valencia['host_id'].astype(str)
Valencia['latitude'] = Valencia['latitude'].astype(str)
Valencia['longitude'] = Valencia['longitude'].astype(str)
Valencia['host_is_superhost_num'] = Valencia['host_is_superhost'].map({'t':1,'f':0})
```

```python
# Tipos de habitación
print(Valencia['room_type'].unique())
print(Valencia['room_type'].value_counts())

# Sub DF
Entire_home = Valencia[Valencia['room_type'] == 'Entire home/apt']
Private_room = Valencia[Valencia['room_type'] == 'Private room']
Hotel_room = Valencia[Valencia['room_type'] == 'Hotel room']
Shared_room = Valencia[Valencia['room_type'] == 'Shared room']
```

```python
#Seleccionar cuantitativas de cada Sub DF
Entire_home_cuant = Entire_home.select_dtypes(['int64','float64'])
Private_room_cuant = Private_room.select_dtypes(['int64','float64'])
Hotel_room_cuant = Hotel_room.select_dtypes(['int64','float64'])
Shared_room_cuant = Shared_room.select_dtypes(['int64','float64'])
```

*Correlacion entire_home*

```python
#Correlación Entire home
Corr_Entire_home = Entire_home_cuant.corr()
Corr_Entire_home

#Absoluto
Corr_Entire_home1= abs(Corr_Entire_home)
Corr_Entire_home1
```

```python
#Correlacion de privaterooms
Corr_Private_room = Private_room_cuant.corr()
Corr_Private_room

#Absoluto
Corr_Private_room1 = abs(Corr_Private_room)
Corr_Private_room1
```

```python
#Correlación Hotel_room
Corr_Hotel_room = Hotel_room_cuant.corr()
Corr_Hotel_room

#Absoluto
Corr_Hotel_room1 = abs(Corr_Hotel_room)
Corr_Hotel_room1
```

```python
#Correlación Shared_room
Corr_Shared_room = Shared_room_cuant.corr()
Corr_Shared_room

#Absoluto
Corr_Shared_room1 = abs(Corr_Shared_room)
Corr_Shared_room1
```

```python
# --- Limpieza robusta y conversión segura ---
Entire_home['host_response_rate'] = (
    Entire_home['host_response_rate']
    .astype(str)                # Convierte todo a texto
    .str.replace('%', '', regex=False)  # Elimina el símbolo %
    .replace('nan', np.nan)     # Reconvierte "nan" textual a valor nulo
    .astype(float)              # Convierte a número flotante
)

Entire_home['host_acceptance_rate'] = (
    Entire_home['host_acceptance_rate']
    .astype(str)
    .str.replace('%', '', regex=False)
    .replace('nan', np.nan)
    .astype(float)
)

# --- Gráfico mejorado ---
plt.figure(figsize=(10,6))
sns.scatterplot(
    x='host_response_rate',
    y='host_acceptance_rate',
    color='blue',
    data=Entire_home,
    alpha=0.7
)

# Línea de tendencia opcional
sns.regplot(
    x='host_response_rate',
    y='host_acceptance_rate',
    data=Entire_home,
    scatter=False,
    color='red',
    line_kws={'lw': 2, 'ls': '--'}
)

plt.xlabel('Host Response Rate (%)', fontsize=12)
plt.ylabel('Host Acceptance Rate (%)', fontsize=12)
plt.title('Relación entre Tasa de Respuesta y Tasa de Aceptación del Anfitrión', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
#Private_room
# --- Limpieza de datos y conversión segura ---
Private_room['host_response_rate'] = (
    Private_room['host_response_rate']
    .astype(str)
    .str.replace('%', '', regex=False)
    .replace('nan', np.nan)
    .astype(float)
)

Private_room['host_acceptance_rate'] = (
    Private_room['host_acceptance_rate']
    .astype(str)
    .str.replace('%', '', regex=False)
    .replace('nan', np.nan)
    .astype(float)
)

# --- Gráfico mejorado ---
plt.figure(figsize=(10,6))  # Aumenta el tamaño del gráfico

sns.scatterplot(
    x='host_response_rate',
    y='host_acceptance_rate',
    color='red',
    data=Private_room,
    alpha=0.7
)

# --- Línea de tendencia (opcional) ---
sns.regplot(
    x='host_response_rate',
    y='host_acceptance_rate',
    data=Private_room,
    scatter=False,
    color='black',
    line_kws={'lw': 2, 'ls': '--'}
)

# --- Etiquetas y formato ---
plt.xlabel('Host Response Rate (%)', fontsize=12)
plt.ylabel('Host Acceptance Rate (%)', fontsize=12)
plt.title('Relación entre Tasa de Respuesta y Tasa de Aceptación (Habitaciones Privadas)', fontsize=14)
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.grid(True, linestyle='--', alpha=0.4)

plt.show()

sns.scatterplot(x= 'host_response_rate', y= 'host_acceptance_rate', color= 'red', data= Private_room)
```

```python
# Variables cuantitativas de Valencia n y obtener su correlación
Valencia_cuant = Valencia.select_dtypes(['int64','float64'])
Valencia_corr = Valencia_cuant.corr()
Valencia_corr
```

```python
# y = review_scores_rating
# x = number_of_reviews_ltm (0.32), estimated_occupancy_l365d (0.43), y  reviews_per_month (0.38).
from sklearn.linear_model import LinearRegression

#Declaramos las variables dependientes e independientes para la regresión lineal
Vars_Indep_r_s_r= Valencia_cuant[['number_of_reviews_ltm', 'estimated_occupancy_l365d','reviews_per_month']]
Var_Dep_r_s_r= Valencia_cuant['review_scores_rating']

# Define modelo
model_r_s_r = LinearRegression()

# Ajustar modelo
model_r_s_r.fit(X= Vars_Indep_r_s_r, y= Var_Dep_r_s_r)

#Verificamos los coeficientes obtenidos para el modelo ajustado
print(model_r_s_r.__dict__)

#Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R Determinación
print(model_r_s_r.score(Vars_Indep_r_s_r,Var_Dep_r_s_r))

#Corroboramos cual es el coeficiente de Determinación de nuestro modelo
coef_Deter_r_s_r =model_r_s_r.score(X=Vars_Indep_r_s_r, y=Var_Dep_r_s_r)
print('R^2: ',coef_Deter_r_s_r)

#Corroboramos cual es el coeficiente de Correlación de nuestro modelo
coef_Correl_r_s_r = np.sqrt(coef_Deter_r_s_r)
print('R: ',coef_Correl_r_s_r)
```

```python
# y = host_acceptance_rate
# x = host_response_rate (0.34), reviews_per_month (0.21), y estimated_revenue_l365d (0.19)

# Clean 'host_response_rate' and 'host_acceptance_rate'
Valencia['host_response_rate'] = Valencia['host_response_rate'].astype(str).str.replace('%', '').astype(float) / 100
Valencia['host_acceptance_rate'] = Valencia['host_acceptance_rate'].astype(str).str.replace('%', '').astype(float) / 100


Vars_Indep_h_a_r = Valencia[['host_response_rate','reviews_per_month','estimated_revenue_l365d']]
Var_Dep_h_a_r = Valencia['host_acceptance_rate']

model_h_a_r = LinearRegression()
model_h_a_r.fit(X=Vars_Indep_h_a_r, y=Var_Dep_h_a_r)

coef_Deter_h_a_r = model_h_a_r.score(Vars_Indep_h_a_r, Var_Dep_h_a_r)
coef_Correl_h_a_r = np.sqrt(coef_Deter_h_a_r)

print("R^2: ", coef_Deter_h_a_r)
print("R: ", coef_Correl_h_a_r)
```
